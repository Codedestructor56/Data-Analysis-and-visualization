```python
import random
import matplotlib.pyplot as plt
x=[0]
y=[0]
def point_saver(l1,l2,num:int):
    if num==0:
        x.append(l1/2)
        y.append(l2/2)
    elif num==1:
        x.append((l1+1)/2)
        y.append(l2/2)
    else:
        x.append((l1+0.5)/2)
        y.append((l2+(3**0.5))/2)
for i in range(50000):
    point_saver(x[-1],y[-1],random.randint(0,2))
plt.plot(x,y,'b+')
plt.show()
```
